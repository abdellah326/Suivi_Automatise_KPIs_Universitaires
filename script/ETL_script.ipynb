```python
# add the Final_Percentage column 
df['Final_Percentage']= df[[
    'Eng1', 'Math1', 'Phy1', 'Chem1', 'Bio1',
    'Eng2', 'Math2', 'Phy2', 'Chem2', 'Bio2',
    'Eng3', 'Math3', 'Phy3', 'Chem3', 'Bio3']
].mean(axis=1)
df
```

```python
# converte student_performance_rating & Internet_Access & gpa_range to number for Scaling
df['Internet_Access_num']= df['has_internet_access'].map({'Yes': 1, 'No':0})
df['gpa_range_num'] = df['gpa_range'].map({'2-2.5': 1, '2.5-3': 2, '3-3.5': 3, '3.5-4': 4})
```

```python
df 
```

```python
def calculer_risque_multicritere(row):
    score_risque = 0
    
    if row['Final_Percentage'] < 55.0:
        score_risque += 3  
    elif row['Final_Percentage'] < 60.0:
        score_risque += 1 
        
    if row['gpa_range_num'] <= 2: 
        score_risque += 2
                
    if row['Internet_Access_num'] == 0:
        score_risque += 1         
   
    if row['family_income'] in ['Less Than 5000', '5000-10000']:
        score_risque += 1
        
    if score_risque >= 4:
        return 1
    else:
        return 0


df['Target_Risk'] = df.apply(calculer_risque_multicritere, axis=1)
df
```

```python
df['student_performance_rating_num'] = df['student_performance_rating'].map({'Satisfactory': 1 , 'Good': 2 ,'Very Good' : 3, 'Excellent': 4})
df['family_income_num'] = df['family_income'].map({'Less Than 5000': 1 , '5000-10000': 2 , '10000-20000': 3, 'More Than 20000':4})
```

```python
print(df.groupby('Target_Risk')[['Final_Percentage']].mean().round(2))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X = df[['Eng1', 'Math1', 'Phy1', 'Chem1', 'Bio1',
    'Eng2', 'Math2', 'Phy2', 'Chem2', 'Bio2',
    'Internet_Access_num','family_income_num' ]]
Y = df['Target_Risk']
X_train, X_test, y_train, y_test = train_test_split(X,Y, test_size=0.2, random_state=42)
rfc = RandomForestClassifier(n_estimators=500, random_state=42,max_depth=15,min_samples_split=5,class_weight='balanced')
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
import xgboost as xgb
clf = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
clf.fit(X_train, y_train)
y_xgbost_pred = clf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_xgbost_pred))
```

```python
importances = pd.Series(rfc.feature_importances_, index=X_train.columns)

plt.figure(figsize=(10, 6))
importances.nlargest(len(X)).sort_values().plot(kind='barh', color='#3498db')

plt.title("Importance des Variables : Facteurs clés pour prédire le risque de décrochage", fontsize=14, fontweight='bold')
plt.xlabel("Niveau d'Importance", fontsize=12)
plt.ylabel("Variables (Features)", fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
importances_xgboost = pd.Series(clf.feature_importances_, index=X_train.columns)

plt.figure(figsize=(10, 6))
importances_xgboost.nlargest(len(X)).sort_values().plot(kind='barh', color='#3498db')

plt.title("Importance des Variables : Facteurs clés pour prédire le risque de décrochage pour xgboost : ", fontsize=14, fontweight='bold')
plt.xlabel("Niveau d'Importance", fontsize=12)
plt.ylabel("Variables (Features)", fontsize=12)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
estimators = [
    ('randomForest', RandomForestClassifier(n_estimators=200, random_state=42, class_weight='balanced')),
    ('xgboost', xgb.XGBClassifier(random_state=42, eval_metric='logloss')),
    ('dicision_tree',DecisionTreeClassifier(criterion='gini', random_state=42))
]

clf_stacking = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression()
)

clf_stacking.fit(X_train, y_train)
y_pred_stack = clf_stacking.predict(X_test)

print("Accuracy of stacking : ", accuracy_score(y_test, y_pred_stack) * 100, "%")
```
